```python
results = pd.merge(races, driver_standings_wins, on='raceId', how='inner')
results.head(2)
```

```python
shifted = results.groupby(level=["driverId"]).shift(1)
shifted
```

```python
results = results.join(shifted.rename(columns=lambda x: x+"_lag"))
results
```

```python

#Drop unnecessary columns
#results.drop(columns=["rank_lag"],inplace=True)
results = results.reset_index()
results = results.fillna(0)
results
```

```python
results = pd.merge(results, drivers, on='driverId', how='left')
```

```python
results
```

```python
results["name"] = results[["forename", "surname"]].apply(" ".join, axis=1)
```

```python
results = results[['name','raceId','year','raceId_lag']]
results.columns = ['name','value','year','lastValue']
results
```

```python
results['lastValue'] = results["lastValue"].astype(int)
results.sort_values(["year","name"],inplace=True)
```

```python
results.head(10)
```

```python
driver_name_last_rec_dict = {}
```

```python
df_new = pd.DataFrame()
```

```python
driver_name_last_rec_dict = {}
df_new = pd.DataFrame()
# Looping through groups
for year_grp, group_df in results.groupby('year'):
    # print(f"Year: {year_grp}")
    if bool(driver_name_last_rec_dict):
        # Find the driver names present in the dict but not in the group dataframe
        drivers_missing_in_curr_df = driver_name_last_rec_dict.keys() - list(group_df['name'])
        # print(f"Missing driver in this year group : {drivers_missing_in_curr_df}")
        for driver_name in drivers_missing_in_curr_df:
            wins = driver_name_last_rec_dict[driver_name]
            df_new = df_new.append({'name':driver_name,'value':wins,'year':year_grp,'lastValue':wins},
                                     ignore_index=True)
    # Update the dictionary with all the new entries
    for index, row in group_df.iterrows():
        driver_name_last_rec_dict[row['name']]=row['value']

    # State of the dictionary
    # print(f"The dict looks like {driver_name_last_rec_dict}")
    
    # concate the newly formed df with this group
    df_new = pd.concat([df_new,group_df])
    # print(f"New Shape of DF: {df_new.shape}")
    # print("\n")

        
```
