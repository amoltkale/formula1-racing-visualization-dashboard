```python
import pandas as pd
import numpy as np
```

```python
drivers = pd.read_csv('archive/drivers.csv')
drivers["driver_name"] = drivers.apply(lambda x:'%s %s' % (x['forename'],x['surname']),axis=1)
drivers
```

```python
circuits = pd.read_csv('../archive/circuits.csv')
circuits
```

```python
races = pd.read_csv('../archive/races.csv')
races
```

```python
races = pd.merge(races, circuits, on='circuitId', how='left')
races
```

```python
#races = races[['raceId','circuitId','circuitRef','name_x','location','country','lat','lng']]
```

```python
crct_count = races.groupby('circuitId')['raceId'].nunique().to_frame("num_races").reset_index()
```

```python
circuits = pd.merge(circuits, crct_count, on='circuitId', how='left')
circuits
```

```python
races = pd.merge(races, crct_count, on='circuitId', how='left')
```

```python
races
```

```python
results = pd.read_csv('../archive/results.csv')
results
```

```python
results = pd.merge(results, races, on='raceId', how='left')
results
```

```python
# Get unique drivers drove on this circuit and constructor participated.
unique_drivers_constructors = results.groupby('circuitId')['driverId','constructorId'].nunique().add_prefix('num_').reset_index()
unique_drivers_constructors = unique_drivers_constructors.rename(columns={'num_driverId':'num_drivers','num_constructorId':'num_constructors'})
unique_drivers_constructors
```

```python
# Get unique drivers drove on this circuit
unique_drivers_winners = results[results['position']=='1'].groupby('circuitId')['driverId'].nunique().reset_index()
unique_drivers_winners = unique_drivers_winners.rename(columns={'driverId':'winners'})
unique_drivers_winners
```

```python
# Get unique pole sitters drove on this circuit
unique_pole_sitters = results[results['grid']==1].groupby('circuitId')['driverId'].nunique().reset_index()
unique_pole_sitters = unique_pole_sitters.rename(columns={'driverId':'pole_sitters'})
unique_pole_sitters
```

```python
# Get the Best Driver per circuit
best_drivers = results[results['position']=='1'].groupby(['circuitId','driverId']).size().to_frame('wins').reset_index() \
 .groupby(['circuitId']).max().reset_index()
best_drivers = pd.merge(best_drivers, drivers, on='driverId', how='left')
best_drivers = best_drivers[['circuitId','driver_name','wins']]
best_drivers
```

```python
circuits = pd.merge(circuits, best_drivers, on='circuitId', how='left')
circuits = pd.merge(circuits, unique_drivers_constructors, on='circuitId', how='left')
circuits = pd.merge(circuits, unique_drivers_winners, on='circuitId', how='left')
circuits = pd.merge(circuits, unique_pole_sitters, on='circuitId', how='left')
circuits
```

```python
circuits.to_csv('circuit_stats_v2.csv',index=False)
```
