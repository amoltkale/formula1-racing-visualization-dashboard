```python
import pandas as pd
```

```python
circuits = pd.read_csv('../archive/circuits.csv')
circuits
```

```python
circuits.groupby('country').count()
```

```python
results = pd.read_csv('../archive/results.csv')
print(f"Shape: {results.shape}")
results.head(2)
```

```python
races = pd.read_csv('../archive/races.csv')
print(f"Shape: {races.shape}")
races.head(2)
```

```python
races = pd.merge(races, circuits, on='circuitId', how='inner')
print(f"Shape: {races.shape}")
races.head(2)
```

```python
results = pd.merge(results, races, on='raceId', how='inner')
print(f"Shape: {results.shape}")
results.head(2)
```

```python
df = results[['year','circuitRef','name_y','country','fastestLapSpeed','driverId']]
```

```python
df.info()
```

```python
df = df[df['fastestLapSpeed']!='\\N']
print(f"Shape: {df.shape}")
df.head(2)
```

```python
df['fastestLapSpeed'] = df.fastestLapSpeed.astype(float)
```

```python
# Grouping by year and the circuit, then picking the records with max. fastest Speed
df = df.loc[df.groupby(['year','country'])['fastestLapSpeed'].idxmax()]
df
```

```python
df['country'].unique().size
```

```python
df['circuitRef'].unique().size
```

```python
df['name_y'].unique().size
```

```python
df = df.rename(columns={'country': 'grand_prix'})
df
```

```python
race_count = df.groupby('grand_prix').size().to_frame('count').reset_index()
```

```python
race_count
```

```python
df = pd.merge(df, race_count, on='grand_prix', how='inner')
df.head(2)
```

```python
df.to_csv('speed_per_circuit.csv',index=False)
```
