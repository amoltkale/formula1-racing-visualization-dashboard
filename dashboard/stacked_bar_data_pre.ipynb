```python
results = pd.merge(results, status, on='statusId', how='left')
results
```

```python
races = pd.read_csv('../archive/races.csv')
races
```

```python
results = pd.merge(results, races, on='raceId', how='left')
results
```

```python
circuits = pd.read_csv('../archive/circuits.csv')
circuits
```

```python
results = pd.merge(results, circuits, on='circuitId', how='left')
results
```

```python
results['status']
```

```python
results.loc[results['status'].str.contains('Lap'), 'status'] = 'Finished'
```

```python
results.loc[results['status'].isin(car_faults), 'status'] = 'Car Fault'
results.loc[results['status'].isin(other_status), 'status'] = 'Other'
results.loc[results['status'].isin(accident_status), 'status'] = 'Accident'
```

```python
results = results[['resultId','year','status','country']]
```
