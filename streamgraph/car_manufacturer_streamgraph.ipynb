```python
import pandas as pd
```

```python
def get_shape(df):
    print("Shape: {0}".format(df.shape))
```

```python
constructor_results = pd.read_csv('../archive/constructor_results.csv')
get_shape(constructor_results)
constructor_results.head(2)
```

```python
races = pd.read_csv('../archive/races.csv')
get_shape(races)
races.head(2)
```

```python
results = pd.merge(constructor_results, races, on='raceId', how='inner')
get_shape(results)
results.head(2)
```

```python
df = results.groupby(['year', 'constructorId'])['points'].sum()
df.to_frame()
```

```python
df = df.reset_index().set_index('year')
df.head(2)
```

```python
df = df.reset_index()
```

```python
constructors = pd.read_csv('../archive/constructors.csv')
get_shape(constructors)
constructors.head(2)
```

```python
results = pd.merge(df, constructors, on='constructorId', how='inner')
get_shape(results)
results.head(2)
```

```python
results = results.set_index('year')
```

```python
results.head(2)
```

```python
results = results.pivot_table(values='points', index=results.index, columns='name', aggfunc='first')
results
```
